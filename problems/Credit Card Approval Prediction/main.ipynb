```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/application_record.csv')
df1.head()
```

```python
df2 = pd.read_csv('./data/credit_record.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, on='ID')
df.head(20)
```

```python
df.dtypes
```

```python
df.STATUS.value_counts()
```

```python
df.STATUS = df.STATUS.replace(['C', 'X'], [6, 7])
df.STATUS = df.STATUS.astype(int)
```

```python
df.hist()
```

```python
df_object = df[df.select_dtypes(include=['object']).columns]
df_object.head()
```

```python
df_numeric = df[df.select_dtypes(include=['int', 'float']).columns]
df_numeric.head()
```

```python
df_object = df_object.fillna('null')
df_object.isna().sum()
```

```python
df_object_encoded = pd.get_dummies(df_object, dtype=int)
df_object_encoded
```

```python
for col in df_object_encoded:
    if 'null' in col:
        df_object_encoded = df_object_encoded.drop(col, axis=1)
df_object_encoded
```

```python
df_object.isna().sum()
```

```python
df_numeric
```

```python
df_numeric.isna().sum()
```

```python
merged_df = pd.concat([df_object_encoded, df_numeric], axis=1)
merged_df
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, KFold
from xgboost import XGBClassifier
```

```python
TARGET = 'STATUS'
X = merged_df.drop(columns=TARGET)
y = merged_df[f'{TARGET}']
```

```python
model = KNeighborsClassifier(n_neighbors=7)
score = cross_val_score(model, X, y, cv=KFold(n_splits=5, shuffle=True))
score.mean(), score
```

```python
model = RandomForestClassifier(n_estimators=100)
score = cross_val_score(model, X, y, cv=KFold(n_splits=5, shuffle=True))
score.mean(), score
```

```python
model = XGBClassifier(n_estimators=1000)
score = cross_val_score(model, X, y, cv=KFold(n_splits=5, shuffle=True))
score.mean(), score
```
